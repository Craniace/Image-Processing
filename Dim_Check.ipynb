```python
# import os
# from PIL import Image

# def check_image_dimensions(folder_name, target_width=227, target_height=227):
#     # Get the current working directory
#     cwd = os.getcwd()

#     # Combine the current working directory with the provided folder name
#     folder_path = os.path.join(cwd, folder_name)

#     # Loop through all files in the folder
#     for filename in os.listdir(folder_path):
#         # Check if the file is an image
#         if filename.endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):
#             # Open the image
#             image_path = os.path.join(folder_path, filename)
#             image = Image.open(image_path)
#             width, height = image.size

#             # Check if the image dimensions match the target dimensions
#             if width != target_width or height != target_height:
#                 print(f"The image {filename} does not have the required dimensions (227x227).")
#                 print(f"Current dimensions: {width}x{height}")
#             else:
#                 print(f"The image {filename} has the correct dimensions (227x227).")


import os
from PIL import Image

def check_image_dimensions(folder_path, target_width=227, target_height=227):
    # Loop through all files in the folder
    for filename in os.listdir(folder_path):
        # Check if the file is an image
        if filename.endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):
            # Open the image
            image_path = os.path.join(folder_path, filename)
            image = Image.open(image_path)
            width, height = image.size

            # Check if the image dimensions match the target dimensions
            if width != target_width or height != target_height:
                print(f"The image {filename} does not have the required dimensions (227x227).")
                print(f"Current dimensions: {width}x{height}")
            else:
                print(f"The image {filename} has the correct dimensions (227x227).")

# Example usage:
folder_path = r"80% of training image with DAUG/new notumor after deleting/"  # Specify the folder path containing images
check_image_dimensions(folder_path)
```

```python
import os

def check_folder_permissions(folder_path):
    if os.access(folder_path, os.R_OK):
        print(f"Read permission is granted on folder: {folder_path}")
    else:
        print(f"Read permission is NOT granted on folder: {folder_path}")

    if os.access(folder_path, os.W_OK):
        print(f"Write permission is granted on folder: {folder_path}")
    else:
        print(f"Write permission is NOT granted on folder: {folder_path}")

    if os.access(folder_path, os.X_OK):
        print(f"Execute permission is granted on folder: {folder_path}")
    else:
        print(f"Execute permission is NOT granted on folder: {folder_path}")

# Example usage
folder_path = "pituitary/"  # Specify the folder path you want to check permissions for
check_folder_permissions(folder_path)
```

```python
def count_images_in_folder(folder_path):
    image_extensions = ('.jpg', '.png', '.jpeg')
    num_images = sum(1 for filename in os.listdir(folder_path) if filename.endswith(image_extensions))
    return num_images
num_images = count_images_in_folder(folder_path)
print(f"Number of images in '{folder_path}': {num_images}")
```
