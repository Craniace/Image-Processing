```python
import os

# folder path
dir_path = r'C:\Users\rappe\OneDrive\Desktop\sharthakimageprocessing\Image-Processing-main\outputfolder\80% of training image\new pituitary after deleting'
count = 0
# Iterate directory
for path in os.listdir(dir_path):
    # check if current path is a file
    if os.path.isfile(os.path.join(dir_path, path)):
        count += 1
print('File count:', count)
todel = count - 5000
print('File to delete:',todel)
import os

def remove_files(directory, num_files_to_remove):
    # List all files in the directory
    files = os.listdir(directory)
    
    # Sort files based on modification time (oldest first)
    files.sort(key=lambda x: os.path.getmtime(os.path.join(directory, x)))
    
    # Remove the specified number of files
    files_to_remove = files[:num_files_to_remove]
    for file_name in files_to_remove:
        file_path = os.path.join(directory, file_name)
        os.remove(file_path)
        print(f"Removed file: {file_name}")

# Example usage
remove_files(dir_path, todel)
```
